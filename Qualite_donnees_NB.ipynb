```python
# valeurs du champ patient_id qui sont dupliques 
ids = df_patient['patient_id']
# Ci-dessous un DataFrame contenant uniquement les lignes correspondantes 
# à des duplications au niveau du champ patient_id.
# Les duplications sont ordonnées de telle manière à afficher en premier 
# les lignes avec éventuellement une valeur manquante au niveau de l'un des champs date_of_birth ou postcode.
df_patient_dup = df_patient[ids.isin(ids[ids.duplicated()])].sort_values(by=['patient_id', 'date_of_birth', 'postcode'], na_position='first')
```

```python
# On visualise ci-dessous l'anomalie des duplications des valeurs de la colonne patient_id,
# au niveau des dix premières lignes. 
df_patient_dup[:10]
```

```python
# Nous avons 403 lignes avec des duplications au niveau du champ patient_id.
# En fait chaque valeur de la colonne patient_id apparait au moins deux fois.
len(df_patient_dup)
```

```python
# Avec la commande suivante, on affiche les dix premières lignes avec des valeurs nulles au niveau de la colonne 'state'
df_patient[df_patient['state'].isnull()][:10]
```

```python
# La présence de valeurs nulles au niveau de champ 'state' est l'une des anomalies de nos données. 
# On dispose de 1990 lignes contenant cette anomalie.
len(df_patient[df_patient['state'].isnull()])
```

```python
# Nous avons ici une source d'anomalies au niveau des champ state'
# Les valeurs dans ce champ ne sont pas normalisées.
# En effet, dans l'exmple suivant :
# pour postcode = 2117, nous avons différentes valeurs correspondantes au niveau du champ 'state'.
df_patient[['postcode', 'state']][df_patient['postcode'] == '2117']
```

```python
# Pour garantir une information fiable au niveau de la colonne state, on se basera donc sur le postcode.
# Au niveau de cette colonne, nous avons également certaines valeurs manquantes
df_patient.loc[:,['postcode', 'state']][df_patient['postcode'].isnull()]
```

```python
# Nous avons 199 valeurs manquantes au niveau du champ postcode
len(df_patient[df_patient['postcode'].isnull()])
```

```python
# L'age présente également une anomalie, car les valeurs de ce champ sont entrées comme constantes.
# On peut facilement voir que ces valeurs ne sont pas correctes, tenant compte de la date de naissance.
df_patient[['date_of_birth', 'age']][:10]
```

```python
# Dans ce contexte aussi, nous allons nous basé sur la colonne date_of_birth, 
# pour calculer dynamiquement et avec exactitude l'age des patients.
df_patient.loc[:,['date_of_birth', 'age']][df_patient['date_of_birth'].isnull()]
```

```python
# Le nombre de lignes représentant cette anomalie vaut 2011
len(df_patient.loc[:,['date_of_birth', 'age']][df_patient['date_of_birth'].isnull()])
```
